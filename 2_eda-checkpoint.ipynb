```python
from nltk import bigrams
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
import string

def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    return text

# Applying the cleaning function to your messages column
data['message'] = data['message'].apply(lambda x: clean_text(str(x)))

# Removing stopwords from the messages
stopwords = set(stopwords.words('english'))
data['message'] = data['message'].apply(lambda x: ' '.join([word for word in str(x).split() if word not in stopwords])) 

channels = data['channel_name'].unique()

for channel in channels:
    channel_data = data[data.channel_name == channel]

    words = [word for word in ' '.join(channel_data['message']).split()]
    bigram_counts = Counter(list(bigrams(words)))

    bigram_df = pd.DataFrame(bigram_counts.most_common(10), columns=['bigram', 'count'])

    # Plotting
    bigram_df.sort_values(by='count', ascending=True).plot(kind='barh', x='bigram', y='count', title=channel)
    plt.show()
```
